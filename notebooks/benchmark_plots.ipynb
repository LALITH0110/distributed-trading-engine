# Scaling Benchmark Plots (DASH-05)

Throughput (orders/sec) and latency (p50/p95/p99) vs. strategy node count (1/2/4/8).

**Data source:** Populated after Phase 10 FABRIC benchmark runs.
Run `scripts/run_benchmarks.py` (Phase 10) to replace placeholder data with real measurements.

Results shown: 3 scenarios × 5 trials each per node count.
- GBM bull run
- BTC/USDT May 2021 crash
- 50K ticks/sec stress


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Replace with real data from Phase 10 benchmark runs
# Shape: {scenario: {node_count: [trial1, trial2, trial3, trial4, trial5]}}

NODE_COUNTS = [1, 2, 4, 8]

# Throughput (orders/sec) — placeholder: linear scaling assumption
throughput_data = {
    "GBM Bull":    {1: [820, 830, 815, 825, 818],
                    2: [1600, 1620, 1590, 1610, 1605],
                    4: [3100, 3080, 3120, 3090, 3110],
                    8: [5800, 5750, 5820, 5780, 5790]},
    "BTC Crash":   {1: [780, 790, 775, 785, 782],
                    2: [1520, 1540, 1510, 1530, 1525],
                    4: [2900, 2880, 2920, 2890, 2905],
                    8: [5200, 5180, 5230, 5190, 5210]},
    "Stress 50K":  {1: [900, 910, 895, 905, 900],
                    2: [1750, 1770, 1740, 1760, 1755],
                    4: [3300, 3280, 3320, 3295, 3310],
                    8: [6100, 6080, 6120, 6090, 6105]},
}

# Latency p99 (microseconds) — placeholder: slight increase with node count (shared RG/ME)
latency_p99_data = {
    "GBM Bull":    {1: [120, 118, 122, 119, 121],
                    2: [135, 132, 138, 134, 136],
                    4: [160, 158, 163, 159, 161],
                    8: [210, 205, 215, 208, 212]},
    "BTC Crash":   {1: [130, 128, 133, 129, 131],
                    2: [148, 145, 151, 147, 149],
                    4: [175, 172, 178, 174, 176],
                    8: [235, 230, 240, 233, 237]},
    "Stress 50K":  {1: [115, 113, 117, 114, 116],
                    2: [128, 125, 131, 127, 129],
                    4: [152, 150, 155, 151, 153],
                    8: [198, 194, 202, 197, 200]},
}

PLACEHOLDER = True  # set to False after Phase 10 data populated
```

```python
fig_tp, ax_tp = plt.subplots(figsize=(8, 4))
colors = plt.cm.tab10.colors
for i, (scenario, data) in enumerate(throughput_data.items()):
    means = [np.mean(data[n]) for n in NODE_COUNTS]
    stds = [np.std(data[n]) for n in NODE_COUNTS]
    ax_tp.errorbar(NODE_COUNTS, means, yerr=stds, marker="o",
                   color=colors[i], label=scenario, capsize=4, linewidth=2)

ax_tp.set_xlabel("Strategy Nodes")
ax_tp.set_ylabel("Throughput (orders/sec)")
ax_tp.set_xticks(NODE_COUNTS)
ax_tp.set_title(f"Throughput Scaling {'[PLACEHOLDER DATA]' if PLACEHOLDER else ''}")
ax_tp.legend()
ax_tp.grid(alpha=0.3)
if PLACEHOLDER:
    ax_tp.text(0.02, 0.98, "PLACEHOLDER — replace with Phase 10 data",
               transform=ax_tp.transAxes, fontsize=9, color="gray",
               va="top", style="italic")
plt.tight_layout()
plt.show()
```

```python
fig_lat_b, ax_lat_b = plt.subplots(figsize=(8, 4))
for i, (scenario, data) in enumerate(latency_p99_data.items()):
    means = [np.mean(data[n]) for n in NODE_COUNTS]
    stds = [np.std(data[n]) for n in NODE_COUNTS]
    ax_lat_b.errorbar(NODE_COUNTS, means, yerr=stds, marker="s",
                       color=colors[i], label=scenario, capsize=4, linewidth=2)

ax_lat_b.set_xlabel("Strategy Nodes")
ax_lat_b.set_ylabel("p99 Latency (\u03bcs)")
ax_lat_b.set_xticks(NODE_COUNTS)
ax_lat_b.set_title(f"p99 Latency vs Node Count {'[PLACEHOLDER DATA]' if PLACEHOLDER else ''}")
ax_lat_b.legend()
ax_lat_b.grid(alpha=0.3)
if PLACEHOLDER:
    ax_lat_b.text(0.02, 0.98, "PLACEHOLDER — replace with Phase 10 data",
                  transform=ax_lat_b.transAxes, fontsize=9, color="gray",
                  va="top", style="italic")
plt.tight_layout()
plt.show()
```
